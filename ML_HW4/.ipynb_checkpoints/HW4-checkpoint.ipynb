```python
# Task 1 code has taken a lot of inspiration from the kmeans.py file provided
#Task 1:
import math
X = np.array([[3,5], [3,4], [2,8], [2,3], [6,2], [6,4], [7,3], [7,4], [8,5], [8,6]])

def man(data, centroid):
    distance = abs(data[0] - centroid[0]) + abs(data[1] - centroid[1])
    return distance

def euc(data, centroid):
    distance = (data[0] - centroid[0])**2 + (data[1] - centroid[1])**2
    return distance

def assign(data, centroids, formula):
    
    if(formula == True):
        d1 = man(data, centroids[0])
        d2 = man(data, centroids[1])
    else:
        d1 = euc(data, centroids[0])
        d2 = euc(data, centroids[1])
    if(d1 < d2):
        return(0)
    else:
        return(1)

def newset(dataset, clusters, k):
    setnew = []
    for i in range(len(dataset)):
        if(clusters[i] == k):
            setnew.append(dataset[i])
    return np.array(setnew)
    
def kmeans(dataset, centroids, formula):
    clusters = []
    for i in range(len(dataset)):
        clusters.append(assign(dataset[i], centroids, formula))
    set1 = newset(dataset, clusters, 0)
    set2 = newset(dataset, clusters, 1)
    print("New centroids are:", np.mean(set1, axis = 0), np.mean(set2, axis = 0))
    
    # Graph the clusters
    x1 = np.array(dataset[:,0])
    x2 = np.array(dataset[:,1])
    clusters = np.array(clusters)
    newcluster1 = np.array(np.mean(set1, axis = 0))
    newcluster2 = np.array(np.mean(set2, axis = 0))
    plt.scatter(newcluster1[0], newcluster1[1], color = 'blue')
    plt.scatter(newcluster2[0], newcluster2[1], color = 'yellow')
    plt.scatter(x1, x2, clusters == 0, color = 'red')
    plt.scatter(x1, x2, clusters == 1, color = 'green')
    plt.show()
    
#1:
centroids = [[4,6], [5,4]]
kmeans(X, centroids, True)

#2: 

centroids = [[4,6], [5,4]]
kmeans(X, centroids, False)

#3:
centroids = [[3,3], [8,3]]
kmeans(X, centroids, True)

#4:
centroids = [[3,2], [4,8]]
kmeans(X, centroids, True)
```

```python
#Task 2:
# This is a modified version of the kmeans code provided
import math
import random
import time
from scipy import spatial
from tkinter import *
from sklearn import datasets
from statistics import mode

#print("hello")
######################################################################
# This section contains functions for loading CSV (comma separated values)
# files and convert them to a dataset of instances.
# Each instance is a tuple of attributes. The entire dataset is a list
# of tuples.
######################################################################

def loadCSV(fileName):
    fileHandler = open(fileName, "rt")
    lines = fileHandler.readlines()
    fileHandler.close()
    del lines[0] # remove the header
    dataset = []
    for line in lines:
        instance = lineToTuple(line)
        dataset.append(instance)
    dataset = dataset[:-1]
    return dataset

# Converts a comma separated string into a tuple
# Parameters
#   line: a string
# Returns: a tuple
def lineToTuple(line):
    # remove leading/trailing witespace and newlines
    cleanLine = line.strip()
    # get rid of quotes
    cleanLine = cleanLine.replace('"', '')
    # separate the fields
    lineList = cleanLine.split(",")
    # convert strings into numbers
    stringsToNumbers(lineList)
    lineTuple = lineList
    return lineTuple

# Destructively converts all the string elements representing numbers
# to floating point numbers.
# Parameters:
#   myList: a list of strings
# Returns None
def stringsToNumbers(myList):
    for i in range(len(myList)):
        if (isValidNumberString(myList[i])) == True:
            myList[i] = float(myList[i])

# Checks if a given string can be safely converted into a positive float.
# Parameters:
#   s: the string to be checked
# Returns: True if the string represents a positive float, False otherwise
def isValidNumberString(s):
    if len(s) == 0:
        return False
    if  len(s) > 1 and s[0] == "-":
          s = s[1:]
    for c in s:
        if c not in "0123456789.":
            return False
    return True

def euc(instance1, instance2):
    if instance1 == None or instance2 == None:
        return float("inf")
    sumOfSquares = 0
    for i in range(0, len(instance1)-1):
        sumOfSquares += (instance1[i] - instance2[i])**2
    return sumOfSquares

def cos(instance1, instance2):
    if instance1 == None or instance2 == None:
        return float("inf")
    #print(instance1[0:4])
    distance = 1 - spatial.distance.cosine(instance1[0:4],instance2[0:4])
    return distance

def jac(instance1, instance2):
    if instance1 == None or instance2 == None:
        return float("inf")
    value1 = 0
    value2 = 0
    for i in range(0, len(instance1)-1):
        value1 += min(instance1[i], instance2[i])
        value2 += max(instance1[i], instance2[i])
    simularity = value1/value2
    return 1-simularity

def meanInstance(name, instanceList):
    numInstances = len(instanceList)
    if (numInstances == 0):
        return
    numAttributes = len(instanceList[0])
    means = [0] * (numAttributes-1) + [name]
    for instance in instanceList:
        for i in range(0, numAttributes-1):
            means[i] += instance[i]
    for i in range(0, numAttributes-1):
        means[i] /= float(numInstances)
    return tuple(means)


def computeCentroids(clusters):
    centroids = []
    for i in range(len(clusters)):
        name = "centroid" + str(i)
        centroid = meanInstance(name, clusters[i])
        centroids.append(centroid)
    return centroids

def assign(instance, centroids, formula):
    #print(instance, centroids[0])
    if(formula == 1):
        minDistance = euc(instance, centroids[0])
        
    elif(formula == 2):
        minDistance = cos(instance, centroids[0])
        
    elif(formula == 3):
        minDistance = jac(instance, centroids[0])
        
    minDistanceIndex = 0
    for i in range(1, len(centroids)):
        if(formula == 1):
            d = euc(instance, centroids[i])
            if (d < minDistance):
                minDistance = d
                minDistanceIndex = i
        if(formula == 2):
            d = euc(instance, centroids[i])
            if (d < minDistance):
                minDistance = d
                minDistanceIndex = i
        if(formula == 3):
            d = jac(instance, centroids[i])
            if (d < minDistance):
                minDistance = d
                minDistanceIndex = i
    return minDistanceIndex

def createEmptyListOfLists(numSubLists):
    myList = []
    for i in range(numSubLists):
        myList.append([])
    return myList

def assignAll(instances, centroids, formula):
    clusters = createEmptyListOfLists(len(centroids))
    for instance in instances:
        clusterIndex = assign(instance, centroids, formula)
        clusters[clusterIndex].append(instance)
    return clusters

def kmeans(instances, k, formula, initCentroids=None):
    result = {}
    if (initCentroids == None or len(initCentroids) < k):
        # randomly select k initial centroids
        random.seed(time.time())
        centroids = random.sample(instances, k)
    else:
        centroids = initCentroids
    prevCentroids = []
    iteration = 0
    while (centroids != prevCentroids):
        #print(centroids, prevCentroids)
        iteration += 1
        clusters = assignAll(instances, centroids, formula)
        prevCentroids = centroids
        centroids = computeCentroids(clusters)
        withinss = computeWithinss(clusters, centroids)
        
    result["clusters"] = clusters
    result["centroids"] = centroids
    result["withinss"] = withinss
    return result

def computeWithinss(clusters, centroids):
    result = 0
    for i in range(len(centroids)):
        centroid = centroids[i]
        cluster = clusters[i]
        for instance in cluster:
            result += euc(centroid, instance)
    return result

def repeatedKMeans(instances, k, n, formula):
    bestClustering = {}
    bestClustering["withinss"] = float("inf")
    for i in range(1, n+1):
        print("k-means trial %d," % i)
        trialClustering = kmeans(instances, k, formula)
        print("withinss: %.1f" % trialClustering["withinss"])
        if trialClustering["withinss"] < bestClustering["withinss"]:
            bestClustering = trialClustering
            minWithinssTrial = i
    print("Trial with minimum withinss:", minWithinssTrial)
    return bestClustering

######################################################################
# Test code
######################################################################

dataset = loadCSV("iris.data")

#Part 1:
#print(dataset)
#label = 3, so k = 3
start = time.time()
first = repeatedKMeans(dataset, 3, 5, 1)
end = time.time()
set1 = first['clusters'][0]
total = len(set1)
set1 = np.array(set1)
set2 = first['clusters'][1]
total += len(set2)
set2 = np.array(set2)
set3 = first['clusters'][2]
total += len(set3)
set3 = np.array(set3)
print("Accuracy:", (np.count_nonzero(set1 == mode(set1[:,4]))+np.count_nonzero(set2 == mode(set2[:,4]))
       + np.count_nonzero(set3 == mode(set3[:,4]))) / total)
print("Time 1 ", end - start)

start = time.time()
second = repeatedKMeans(dataset, 3, 5, 2)
end = time.time()
print("Time 2 ", end - start)

set1 = second['clusters'][0]
total = len(set1)
set1 = np.array(set1)
set2 = second['clusters'][1]
total += len(set2)
set2 = np.array(set2)
set3 = second['clusters'][2]
total += len(set3)
set3 = np.array(set3)
print("Accuracy:", (np.count_nonzero(set1 == mode(set1[:,4]))+np.count_nonzero(set2 == mode(set2[:,4]))
       + np.count_nonzero(set3 == mode(set3[:,4]))) / total)

start = time.time()
third = repeatedKMeans(dataset, 3, 5, 3)
end = time.time()
print("Time 3 ", end - start)

set1 = third['clusters'][0]
total = len(set1)
#print(len(set1))
set1 = np.array(set1)
set2 = third['clusters'][1]
total += len(set2)
#print(len(set2))
set2 = np.array(set2)
set3 = third['clusters'][2]
total += len(set3)
#print(len(set3))
#print(total)
set3 = np.array(set3)
print("Accuracy:", (np.count_nonzero(set1 == mode(set1[:,4]))+np.count_nonzero(set2 == mode(set2[:,4]))
       + np.count_nonzero(set3 == mode(set3[:,4]))) / 149)
```
